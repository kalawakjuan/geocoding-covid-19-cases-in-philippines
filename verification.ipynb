```python
import pandas as pd
```

```python
dframe = pd.read_csv('DOH COVID Data Drop_ 20220917 - 04 Case Information_batch_3.csv')
```

```python
health_status = dframe[(dframe.HealthStatus != 'RECOVERED') & (dframe.HealthStatus != 'DIED')]
health_status
```

```python
health_status_no_nul = health_status[health_status.BarangayPSGC.notnull()].groupby('BarangayPSGC').agg(active_cases = ('BarangayPSGC', 'count'))
health_status_no_nul
```

```python
zeroes = len(health_status) - len(health_status_no_nul)
```

```python
MILD = health_status_no_nul[(health_status_no_nul.HealthStatus == 'MILD')].groupby('BarangayPSGC', as_index=False).agg(mild = ('HealthStatus', 'count')).sort_values('mild', ascending=False)
CRITICAL = health_status_no_nul[(health_status_no_nul.HealthStatus == 'CRITICAL')].groupby('BarangayPSGC', as_index=False).agg(critical = ('HealthStatus', 'count')).sort_values('critical', ascending=False)
SEVERE = health_status_no_nul[(health_status_no_nul.HealthStatus == 'SEVERE')].groupby('BarangayPSGC', as_index=False).agg(severe = ('HealthStatus', 'count')).sort_values('severe', ascending=False)
ASYMPTOMATIC = health_status_no_nul[(health_status_no_nul.HealthStatus == 'ASYMPTOMATIC')].groupby('BarangayPSGC', as_index=False).agg(asymptomatic = ('HealthStatus', 'count')).sort_values('asymptomatic', ascending=False)
MODERATE = health_status_no_nul[(health_status_no_nul.HealthStatus == 'MODERATE')].groupby('BarangayPSGC', as_index=False).agg(moderate = ('HealthStatus', 'count')).sort_values('moderate', ascending=False)
```

```python
total = len(ASYMPTOMATIC) + len(MILD) + len(MODERATE) + len(CRITICAL) + len(SEVERE)
```

```python
total_total = total - zeroes
```
