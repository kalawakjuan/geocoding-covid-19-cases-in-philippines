```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('DOH COVID Data Drop_ 20220917 - 04 Case Information_batch_3.csv')
df.head()
```

```python
health_status = df[(df.HealthStatus != 'RECOVERED') & (df.HealthStatus != 'DIED')]
health_status
```

```python
MILD = health_status[(health_status.HealthStatus == 'MILD')].groupby('BarangayPSGC', as_index=False).agg(mild = ('HealthStatus', 'count')).sort_values('mild', ascending=False)
CRITICAL = health_status[(health_status.HealthStatus == 'CRITICAL')].groupby('BarangayPSGC', as_index=False).agg(critical = ('HealthStatus', 'count')).sort_values('critical', ascending=False)
SEVERE = health_status[(health_status.HealthStatus == 'SEVERE')].groupby('BarangayPSGC', as_index=False).agg(severe = ('HealthStatus', 'count')).sort_values('severe', ascending=False)
ASYMPTOMATIC = health_status[(health_status.HealthStatus == 'ASYMPTOMATIC')].groupby('BarangayPSGC', as_index=False).agg(asymptomatic = ('HealthStatus', 'count')).sort_values('asymptomatic', ascending=False)
MODERATE = health_status[(health_status.HealthStatus == 'MODERATE')].groupby('BarangayPSGC', as_index=False).agg(moderate = ('HealthStatus', 'count')).sort_values('moderate', ascending=False)
```

```python
summary = ASYMPTOMATIC
summary = summary.merge(MILD, how='outer', on='BarangayPSGC')
summary = summary.merge(MODERATE, how='outer', on='BarangayPSGC')
summary = summary.merge(CRITICAL, how='outer', on='BarangayPSGC')
summary = summary.merge(SEVERE, how='outer', on='BarangayPSGC')
summary
```

```python
gdf = gpd.read_file('Barangays.shp')
gdf
```

```python
gpd_summary = gpd.GeoDataFrame(summary)
```

```python
merged_data = gpd.GeoDataFrame.merge(gpd_summary, gdf, how='right', left_on='BarangayPSGC', right_on='ADM4_PCODE')

merged_data.set_crs('epsg:4326')

merged_data.fillna(0, inplace=True)
```
